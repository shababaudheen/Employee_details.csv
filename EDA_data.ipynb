```python
Team_frequency = data1.Team.value_counts() 
Team_percent = (Team_frequency / len(data1)) * 100 
Team = pd.DataFrame({"Count": Team_frequency, "Percentage": Team_percent}) 
Team
```

#### Data Story

```python
print(data1.Position.value_counts()) 
sns.barplot(x = ['SG', 'PF', 'PG', 'SF', 'C'], y = data1.Position.value_counts(), color = 'purple', edgecolor = 'black', width = 0.4, zorder = 3) 
plt.grid(zorder = 0) 
plt.show()
```

#### Data Story 
Segregating the employees w.r.t. the positions. 
***

```python
data1.Age.value_counts()
```

#### Data Story 
The above table shows number of people having a particular age. 
***

```python
plt.hist(data1['Age'], bins = [18, 21, 24, 27, 30, 33, 36, 39, 42], edgecolor = 'red', color = 'teal') 
plt.title('Frequency distribution of Age')
plt.xlabel('Age') 
plt.ylabel('Frequency') 
plt.ylim(0, 140)
plt.xlim(15, 45)

plt.show()
```

#### Data Story 
The graph shows different age group of employees. 
***

```python
data1[data1['Salary'] == data1['Salary'].max()]
```
